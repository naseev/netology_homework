# Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. <br> 
Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
movie_id = joined[joined.rating == 5.0][['movieId','rating']].groupby('movieId').count().sort_values('rating', ascending = False).head(1)
```

```python
"{} {} {} {}".format('Фильм', 
                     list(joined[joined.movieId == list(movie_id.index)]['title'].unique()),
                     'имеет больше всего оценок 5.0:',
                     movie_id['rating'].iloc[0])
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. <br> Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('Desktop/power.csv')
data.head()
```

```python
pribaltika = data['country'].isin(['Latvia', 'Estonia', 'Lithuania'])
years = (data['year'] == 2005) | (data['year'] == 2006) | (data['year'] == 2007) | (data['year'] == 2008) | (data['year'] == 2009) | (data['year'] == 2010)
category = (data['category'] == 4) | (data['category'] == 12) | (data['category'] == 21)
quantity = data['quantity'] > 0
```

```python
data[pribaltika & years & category & quantity][['country', 'year', 'quantity']].groupby(['country', 'year']).sum()
```

```python
"{} {}".format('суммарное потребление стран Прибалтики категорий 4, 12 и 21 за период с 2005-10 гг.:',
               data[pribaltika & years & category & quantity]['quantity'].sum())
```
