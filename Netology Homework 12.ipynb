# Задание 1
Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

1. если источник traffic_source равен yandex или google, то в source_type ставится organic
2. для источников paid и email из России - ставим ad
3. для источников paid и email не из России - ставим other
4. все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
logs = pd.read_csv('Desktop/visit_log.csv', sep = ';')
logs.head()
```

```python
# 1-ый пункт задания
logs.loc[(logs.traffic_source == 'yandex') | (logs.traffic_source == 'google'), 'source_type'] = 'organic'
logs.head()
```

```python
# 2-ой пункт задания
logs.loc[((logs.traffic_source == 'paid') | (logs.traffic_source == 'email')) & (logs.region == 'Russia'), 'source_type'] = 'ad'
logs[logs.traffic_source == 'paid'].head()
```

```python
# 3-ий пункт задания
logs.loc[((logs.traffic_source == 'paid') | (logs.traffic_source == 'email')) & (logs.region != 'Russia'), 'source_type'] = 'other'
logs[logs.traffic_source == 'email'].head()
```

```python
# 4-ый пункт задания
logs['source_type'] = logs['source_type'].fillna(logs['traffic_source'])
logs.head()
```

# Задание 3
В датафрейме data создайте столбец lemmas, в котором вычислите леммы поисковых запросов из столбца keyword. Леммы должны иметь строковый тип.

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
```

```python
from pymystem3 import Mystem
```

```python
m = Mystem()
lemmas_list = []

for search in data['keyword']:
    lemmas = m.lemmatize(search)
    lemmas = ' '.join(lemmas)
    lemmas_list.append(str(lemmas))
```

```python
lemmas_list
```

```python
data['lemmas'] = lemmas_list
```
