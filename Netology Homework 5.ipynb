#1 Задание

````python
import json
````

````python
my_dict = {}

with open('purchase_log.txt') as f:
    for line in f:
        line = line.strip()
        
        purchases = json.loads(line)
        my_dict[purchases['user_id']] = purchases['category']
````

````python
del my_dict['user_id']
````

````python
my_dict
````

#2 Задание

````python
import csv

visits = []

with open('visit_log.csv','rt') as f:
    data = csv.reader(f)
    for row in data:
        visits.append(row)
````

````python
for i in range(0,len(visits)):
    try:
        visits[i].append(my_dict[visits[i][0]])
    except:
        visits[i].append('no purchase')
````

````python
del visits[0]
````

````python
visits
````

#3 Задание

````python
funnel = []

for i in range(0,len(visits)): 
    while visits[i][2] != 'no purchase':
        funnel.append(visits[i])
        break
````

````python
funnel
````

````python
import csv

with open("funnel.csv", "w", newline = '') as f:
    writer = csv.writer(f)
    writer.writerow(['user_id', 'source', 'category'])
    writer.writerows(funnel)
````
